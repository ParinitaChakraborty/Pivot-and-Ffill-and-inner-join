```python
import pandas as pd
import numpy as np

df = pd.read_excel(r'/Users/parinita/Downloads/test_data.xlsx',sheet_name='Sheet1')
df['row_num'] = np.arange(len(df))
print (df)
```

```python
out = df.pivot_table('Header2', ['row_num'], 'Header1')
R1 = out.add_prefix('A_')
R1
```

```python
val_final = out.fillna(method='ffill')
R2 = val_final.add_prefix('B_')
R2
```

```python
result = pd.concat([R1, R2], axis=1, join="inner")
result
```
